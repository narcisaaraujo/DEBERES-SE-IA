```python
from neo4j import GraphDatabase

class Neo4jService(object):

    def __init__(self, uri, user, password):
        self._driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self._driver.close()      
    
    def crear_candidato(self, tx, nombre):
        tx.run("CREATE (:Candidato {nombre: $nombre})", nombre=nombre)

    def crear_twitter(self, tx, nombre):
        tx.run("CREATE (:Twitter {nombre: $nombre})", nombre=nombre)
    
    def crear_usuario(self, tx, nombre):
        tx.run("CREATE (:Usuario {nombre: $nombre})", nombre=nombre)
        
    def crear_fecha(self, tx, nombre):
        tx.run("CREATE (:Fecha {nombre: $nombre})", nombre=nombre)
        
    def crear_contenido(self, tx, nombre):
        tx.run("CREATE (:Contenido {nombre: $nombre})", nombre=nombre)
        
    def crear_retweet(self, tx, nombre):
        tx.run("CREATE (:Retweet {nombre: $nombre})", nombre=nombre)
        
    def crear_dispositivo(self, tx, nombre):
        tx.run("CREATE (:Dispositivo {nombre: $nombre})", nombre=nombre)
        
    def crear_descripcion(self, tx, nombre):
        tx.run("CREATE (:Descripcion {nombre: $nombre})", nombre=nombre)
    
    def crear_ubicacion(self, tx, nombre):
        tx.run("CREATE (:Ubicacion {nombre: $nombre})", nombre=nombre)
    
    def crear_seguidores(self, tx, nombre):
        tx.run("CREATE (:Seguidores {nombre: $nombre})", nombre=nombre)
        
    
    def crear_amigos(self, tx, nombre):
        tx.run("CREATE (:Amigos {nombre: $nombre})", nombre=nombre)
    
    
    
    
    def crear_relacion_usuario_amigos(self, tx, usuario, amigos):
        tx.run("MATCH (a:Usuario {nombre: $usuario}) "
               "MATCH (b:Amigos {nombre: $amigos}) "
               "MERGE (a)-[:_amigos_]->(b)",
               usuario=usuario, amigos=amigos)
        
    def crear_relacion_usuario_seguidores(self, tx, usuario, seguidores):
        tx.run("MATCH (a:Usuario {nombre: $usuario}) "
               "MATCH (b:Seguidores {nombre: $seguidores}) "
               "MERGE (a)-[:_seguidores_]->(b)",
               usuario=usuario, seguidores=seguidores)   
        
    def crear_relacion_contenido_fecha(self, tx, contenido, fecha):
        tx.run("MATCH (a:Contenido {nombre: $contenido}) "
               "MATCH (b:Fecha {nombre: $fecha}) "
               "MERGE (a)-[:_Fecha_]->(b)",
               contenido=contenido, fecha=fecha) 
        
    def crear_relacion_contenido(self, tx, twitter, contenido):
        tx.run("MATCH (a:Twitter {nombre: $twitter}) "
               "MATCH (b:Contenido {nombre: $contenido}) "
               "MERGE (a)-[:_contenido_]->(b)",
               twitter=twitter, contenido=contenido) 
        
    def crear_relacion_usuario(self, tx, usuario, user_nombre):
        tx.run("MATCH (a:Usuario {nombre: $usuario}) "
               "MATCH (b:Usuario {nombre: $user_nombre}) "
               "MERGE (a)-[:_Usuario_]->(b)",
               usuario=usuario, user_nombre=user_nombre) 
        
    def crear_relacion_retweet(self, tx, contenido, retweet):
        tx.run("MATCH (a:Contenido {nombre: $contenido}) "
               "MATCH (b:Retweet {nombre: $retweet}) "
               "MERGE (a)-[:_Retweet_]->(b)",
               contenido=contenido, retweet=retweet)
        
    def crear_relacion_dispositivo(self, tx, contenido, dispositivo):
        tx.run("MATCH (a:Contenido {nombre: $contenido}) "
               "MATCH (b:Dispositivo {nombre: $dispositivo}) "
               "MERGE (a)-[:_DESDE_]->(b)",
               contenido=contenido, dispositivo=dispositivo)
        
    def crear_relacion_descripcion(self, tx, usuario_nombre, descripcion):
        tx.run("MATCH (a:Usuario {nombre: $usuario_nombre}) "
               "MATCH (b:Descripcion {nombre: $descripcion}) "
               "MERGE (a)-[:_descripcion_]->(b)",
               usuario_nombre=usuario_nombre, descripcion=descripcion)
        
        
    def crear_relacion_usuario_ubicacion(self, tx, usuario, ubicacion):
        tx.run("MATCH (a:Usuario {nombre: $usuario}) "
               "MATCH (b:Ubicacion {nombre: $ubicacion}) "
               "MERGE (a)-[:_ubicacion_]->(b)",
               usuario=usuario, ubicacion=ubicacion) 
        
        
        
        
        
    def crear_relacion_candidato_usuario(self, tx, nombre_candidato, nombre_usuario):
        tx.run("MATCH (a:Candidato {nombre: $nombre_candidato}) "
               "MATCH (b:Usuario {nombre: $nombre_usuario}) "
               "MERGE (a)-[:candidato_usuario]->(b)",
               nombre_candidato=nombre_candidato, nombre_usuario=nombre_usuario)
        
    def crear_relacion_candidato_twitter(self, tx, nombre_candidato, twitter):
        tx.run("MATCH (a:Candidato {nombre: $nombre_candidato}) "
               "MATCH (b:Twitter {nombre: $twitter}) "
               "MERGE (a)-[:candidato_twitter]->(b)",
               nombre_candidato=nombre_candidato, twitter=twitter)
        
    def crear_relacion_usuario_twitter(self, tx, usuario, twitter):
        tx.run("MATCH (a:Usuario {nombre: $usuario}) "
               "MATCH (b:Twitter {nombre: $twitter}) "
               "MERGE (a)-[:_User_Twitter_]->(b)",
               usuario=usuario, twitter=twitter)

print('paso')
   
```
