```python
from neo4j import GraphDatabase

class Neo4jService(object):

    def __init__(self, uri, user, password):
        self._driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self._driver.close()      
    
    def crear_cabecera(self, tx, nombre):
        tx.run("CREATE (:Cabecera {nombre: $nombre})", nombre=nombre)
    
    def crear_candidato(self, tx, nombre):
        tx.run("CREATE (:Candidato {nombre: $nombre})", nombre=nombre)
        
    def crear_seguidores(self, tx, nombre):
        tx.run("CREATE (:Seguidores {nombre: $nombre})", nombre=nombre)

    def crear_tweets(self, tx, nombre):
        tx.run("CREATE (:Tweets {nombre: $nombre})", nombre=nombre)
    
    def crear_retweet(self, tx, nombre):
        tx.run("CREATE (:Retweet {nombre: $nombre})", nombre=nombre)
    
#    def crear_links(self, tx, nombre):
#        tx.run("CREATE (:Links {nombre: $nombre})", nombre=nombre)
        
    def crear_likes(self, tx, nombre):
        tx.run("CREATE (:Likes {nombre: $nombre})", nombre=nombre)
        
    def crear_usuario(self, tx, nombre):
        tx.run("CREATE (:Usuario {nombre: $nombre})", nombre=nombre)

    def crear_fecha(self, tx, nombre):
        tx.run("CREATE (:Fecha {nombre: $nombre})", nombre=nombre)
    
    def crear_contenido(self, tx, nombre):
        tx.run("CREATE (:Contenido {nombre: $nombre})", nombre=nombre)        
   
       
    
    def crear_relacion_cabeceras(self, tx, nombre_cabecera, nombre_usuario):
        tx.run("MATCH (a:Cabecera {nombre: $nombre_cabecera}) "
               "MATCH (b:Usuario {nombre: $nombre_usuario}) "
               "MERGE (a)-[:Usuario]->(b)",
               nombre_cabecera=nombre_cabecera, nombre_usuario=nombre_usuario)
    
    def crear_relacion_tweet(self, tx, nombre_tweets, nombre_tweet):
        tx.run("MATCH (a:Tweets {nombre: $nombre_tweets}) "
               "MATCH (b:Contenido {nombre: $nombre_tweet}) "
               "MERGE (a)-[:Tweets]->(b)",
               nombre_tweets=nombre_tweets, nombre_tweet=nombre_tweet)
    
    
    
    def crear_relacion_fecha(self, tx, nombre_tweets, nombre_fecha):
        tx.run("MATCH (a:Tweets {nombre: $nombre_tweets}) "
               "MATCH (b:Fecha {nombre: $nombre_fecha}) "
               "MERGE (a)-[:Fecha]->(b)",
               nombre_tweets=nombre_tweets, nombre_fecha=nombre_fecha)

    #relacion de candidato   cabecera links noticias 
    def crear_relacion_candidato_usuario(self, tx, nombre_candidato, nombre_usuario):
        tx.run("MATCH (a:Candidato {nombre: $nombre_candidato}) "
               "MATCH (b:Cabecera {nombre: $nombre_usuario}) "
               "MERGE (a)-[:Candidato_usuario]->(b)",
               nombre_candidato=nombre_candidato, nombre_usuario=nombre_usuario)
        
    def crear_relacion_candidato_fecha(self, tx, nombre_candidato, nombre_fecha):
        tx.run("MATCH (a:Candidato {nombre: $nombre_candidato}) "
               "MATCH (b:Tweets {nombre: $nombre_fecha}) "
               "MERGE (a)-[:Candidato_Fecha]->(b)",
               nombre_candidato=nombre_candidato, nombre_fecha=nombre_fecha)
    
    def crear_relacion_candidato_contenido(self, tx, nombre_candidato, nombre_tweets):
        tx.run("MATCH (a:Candidato {nombre: $nombre_candidato}) "
               "MATCH (b:Tweets {nombre: $nombre_tweets}) "
               "MERGE (a)-[:Candidato_Tweets]->(b)",
               nombre_candidato=nombre_candidato, nombre_tweets=nombre_tweets)
        
print("ok")
```

```python
# sacar todos los tweets del candidato
data = []
tweet_ids = set()
last_position = driver.execute_script("return window.pageYOffset;")
scrolling = True

while scrolling:
    page_cards = driver.find_elements_by_xpath('//div[@data-testid="tweet"]')
    for card in page_cards[-15:]:
        tweet = get_tweet_data(card)
        if tweet:
            tweet_id = ''.join(tweet)
            if tweet_id not in tweet_ids:
                tweet_ids.add(tweet_id)
                data.append(tweet)
            
    scroll_attempt = 0
    while True:
        # check scroll position
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        sleep(2)
        curr_position = driver.execute_script("return window.pageYOffset;")
        if last_position == curr_position:
            scroll_attempt += 1
            
            # end of scroll region
            if scroll_attempt >= 3:
                scrolling = False
                break
            else:
                sleep(2) # attempt another scroll
        else:
            last_position = curr_position
            break
```
